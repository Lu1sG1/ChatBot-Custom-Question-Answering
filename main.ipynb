```python
with open(r"QA_project/train.json", "r") as read_file:
    train = json.load(read_file)
```

```python
with open(r"QA_project/test.json", "r") as read_file:
    test = json.load(read_file)
```

```python
test
```

```python
model_type="bert"
model_name= "bert-base-cased"
```

```python
### Advanced Methodology
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "use_cached_eval_features": True,
    "output_dir": f"QA_project/outputs/{model_type}",
    "best_model_dir": f"QA_project/outputs/{model_type}/best_model",
    "evaluate_during_training":True,
    "max_seq_length": 128,
    "num_train_epochs": 50,
    "evaluate_during_training_steps": 1000,
    "wandb_project": "Question Answer 3 Application",
    "wandb_kwargs": {"name": model_name},
    "save_model_every_epoch": False,
    "save_eval_checkpoints": False,
    "n_best_size":3,
    "use_early_stopping": False,
    #"early_stopping_metric": "mcc",
     "n_gpu": 0,
    # "manual_seed": 4,
    # "use_multiprocessing": False,
    "train_batch_size": 64,
    "eval_batch_size": 64,
    # "config": {
    #     "output_hidden_states": True
    # }
}
     
```

```python
model = QuestionAnsweringModel(
    model_type,model_name, args=train_args,use_cuda=True
)
```

```python
# Train the model
model.train_model(train, eval_data=test)
```

```python
# Evaluate the model
result, texts = model.eval_model(test)
```

```python
result
```
